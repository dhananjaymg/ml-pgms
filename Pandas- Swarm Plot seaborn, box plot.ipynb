```python
#importing all necessary libs 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as st1
import statistics as stats2
from IPython.display import Image, display
```

```python
#load the CSV files 
iris_df = pd.read_csv('iris.data.csv')
#iris_df

states_df = pd.read_csv('2008_swing_states.csv')
#states_df

iris_df.head(5)
```

```python
plt.hist(states_df['dem_share'])
plt.xlabel('Total Votes')
plt.ylabel('States')
plt.show()
```

```python
petal_length = iris_df['petal_length']
#petal_length
n_data=len(petal_length)
n_bins=np.sqrt(n_data)
n_bins=int(n_bins)
plt.hist(petal_length, n_bins)
plt.xlabel('Total Votes')
plt.ylabel('States')
plt.show()
```
