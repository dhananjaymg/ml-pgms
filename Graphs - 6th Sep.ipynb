```python
import matplotlib.pyplot as plt
```

```python
x=[1,3,5,7,9]  #x is 5 elem
y=[3,5,6.5,8.5,9] #y also should be same as x elm i.e 5
plt.plot(x,y) 
#plt.xlabel("X level") 
#plt.ylabel("Y level") 
plt.xlabel("Year of Experience") 
plt.ylabel("Salary in Lakhas") 
plt.title("Sal vs Exp")
plt.show()
```

```python
#Graph with Legend using Label as properties for X and Y labels
x1=[1,3,5] 
y1=[2,4,1] 
plt.plot(x1,y1,label="First Line")  #using label properties we show the legend on the graph
x2=[10,13,16] 
y2=[12,6,10] 
plt.plot(x2,y2,label="Second Line") 

plt.xlabel("X level") 
plt.ylabel("Y level") 
plt.legend() #this is used to render legend on the grap Right Top Corner
plt.title("New Graph \nToday's Graph") #\n - new line 
plt.show()
```

```python
#add new line to the Table heading
x1=[2,4,6,8] 
y1=[1,3,5,7] 
plt.bar(x1,y1) # to display in bar chart we use .bar() method of Plot lib

plt.xlabel("X level") 
plt.ylabel("Y level") 
plt.title("New Graph \nToday's Graph") #\n - new line 
plt.show()
```

```python
#two graphs side by side display
x1=[2,4,6,8] 
y1=[1,3,5,7] 


x2=[1,3,5,7] 
y2=[5,6,3,9] 


plt.bar(x2,y2) 
plt.bar(x1,y1) 
plt.xlabel("X level") 
plt.ylabel("Y level") 
plt.title("New Graph \nToday's Graph") #\n - new line 
#plt.legend() 
plt.show()
```

```python
#changing the bar color 

x1=[2,4,6,8] 
y1=[1,3,5,7] 
plt.bar(x1,y1,label='bar1',color='r') 

x2=[1,3,5,7] 
y2=[5,6,3,9] 
plt.bar(x2,y2,label='bar2',color='c') 

plt.xlabel("X level") 
plt.ylabel("Y level") 
plt.title("New Graph \nToday's Graph") #\n - new line 
plt.legend() 
plt.show()
```

```python
# trying to do a comparision using population and bins for frequency macthing result and how it is disturibute 

population=[4,20,15,23,48,45,48,49,60,65,54,71,78,82,103,120] 
bins=[0,20,40,60,80,100,120] 
plt.hist(population,bins,histtype='bar',rwidth=0.5,label='population') 
plt.xlabel("X level") 
plt.ylabel("Y level") 
plt.title("New Graph \nToday's Graph") #\n - new line 
plt.legend() 
plt.show()
```

```python
#stack Plot 

days=[1,2,3,4] 
sleeping=[8,9,7,6] 
playing=[2,3,1,6] 
study=[4,6,9,5] 
working=[10,6,7,8] 
plt.stackplot(days,sleeping,playing,study,working,colors=['m','c','r','k']) 
plt.xlabel("days") 
plt.ylabel("hours") 
plt.title("New Graph \nToday's Graph") #\n - new line 
#plt.legend() 
plt.show()
```

```python
# passign the legend to each of stack, using plot
days=[1,2,3,4] 
sleeping=[8,9,7,6] 
playing=[2,3,1,6] 
study=[4,6,9,4] 
working=[10,6,7,8] 
plt.plot([],[],label='sleeping',color='m',linewidth=2) 
plt.plot([],[],label='playing',color='c',linewidth=2) 
plt.plot([],[],label='study',color='r',linewidth=2) 
plt.plot([],[],label='working',color='k',linewidth=2) 
plt.stackplot(days,sleeping,playing,study,working,colors=['m','c','r','k']) 
plt.xlabel("X level - Day") 
plt.ylabel("Y level -Hours") 
plt.title("New Graph \nToday's Graph") #\n - new line 
plt.legend() 
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise: 
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs' 
sizes = [15, 30, 45, 10] 
explode = (0, 0.1, 0, 0) # only "explode" the 2nd slice (i.e. 'Hogs') 

fig1, ax1 = plt.subplots() # we need to get the axis to display from subplots method
ax1.pie(sizes, explode=explode, labels=labels, autopct='%2.1f%%', 
shadow=50, startangle=180) 
ax1.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle. 

plt.show()
```

```python
#stackedBar plot using plot.bar


A = [5., 30., 45., 22.] 
B = [5., 25., 50., 20.] 

X = range(4) 

plt.bar(X, A, color = 'b' ,label='A') 
plt.bar(X, B, color = 'r', bottom = A, label='B') 
plt.xlabel("X level ") 
plt.ylabel("Y level -A&B") 
plt.legend() 
plt.show()
```
